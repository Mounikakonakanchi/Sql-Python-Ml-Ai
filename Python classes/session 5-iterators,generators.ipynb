Iterators and Generators

An iterator is an object that lets you go through items one at a time. <br>

It remembers where you left off, so every time you ask for the next value, it continues from there.<br>

An iterator is something you can loop through but the data is already available

In Python, an iterator must have two methods: <br>

__iter__() → returns the iterator object <br> 

__next__() → returns the next value 

Use Iterator when :<br>
You want to loop manually using next() <br>
You're working with built-in iterator objects <br>
You want complete iteration control

Example : Phone photo gallery one by one By using iterator nothing new will be generated  <br>
Pages in the book we can access the next page

```python
colors = ["red", "green", "blue"]
it = iter(colors)
for color in it:
    print(color)
```

```python
a = [1, 2, 3]
it = iter(a)           # create iterator
print(next(it))   # 1
print(next(it))   # 2
print(next(it))   # 3
```

```python
name = "Mouni"
it = iter(name)
print(next(it))   # M
print(next(it))   # o
print(next(it))   # u
print(next(it))   #n
```

Using loop with iterator

```python
colors = ["red", "green", "blue"]
it = iter(colors)
for color in it:
    print(color)
```

Generator : A geneartor is a special type of Iterator that is written like a function but uses yield keyword instead of return <br>
It pauses when it yields, and resumes from where it left off <br>
it is used to generate values on fly , saving memory <br>
Return end the function and sends back a value <br>
Yield pauses the function and remembers its state , allowing it to resume later

When to use Generator : <br>
You need values one-by-one <br>
Data is too large and infinite <br>
You want simple & efficient code

Examples: Instagram scrolling and when refreshed it generates new posts <br>
Ticket Dispenser always generates new ticket following the previous


```python
# Generator function for the cube of numbers 
def gencubes(n):
    for num in range(n):
     yield num**3
```

```python
gencubes(10)
```

```python
for x in gencubes (9):
    print(x)
```

```python
def gen_numbers(n):
    for i in range(n):
        yield i              
for num in gen_numbers(5):    # Using the generator
    print(num)
```

Generator for Fibonacci Numbers

```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
for num in fib(10):
    print(num)
```

Map() : is a tool in Python that takes a function and a list and applies that function to every item in the list, producing a new sequence of results.<br>

if You have many items and wanted to do the same action to each item map() does this automatically and gives you the transformed items.<br>

Here the original list will not changed.<br>

The result is a new sequence (can be converted to a list, tuple, etc.).<br>

It’s useful when you want to process each item the same way without writing a loop.<br>

Map transforms eaxh item individually

Exmaples: Conversion of temperature , apply discount to all prices 
       

```python
def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9
```

```python
temperatures_f = [32, 68, 77, 104]
temperatures_c = map(fahrenheit_to_celsius, temperatures_f)
print(list(temperatures_c))  # Convert map object to list and print
```

Here we can use lambda so that no need to define and name fahrenheit and celsius functions

```python
names = ["mouni", "sai", "bhagi"]
capital = list(map(lambda name: name.title(), names))

print(capital)
```

Reduce : reduce() is used to combine all items in a list into a single value. <br>
reduce() starts with the first two items, combines them, then takes the result and the next item, combines again, and so on <br>

Always returns one single value.<br>

The function you use must take two arguments.<br>

Useful for sum, product, max, or any cumulative operation.

Examples : Total bill , sum of numbers , maximum of numbers , whenever the output in single value is needed

```python
from functools import reduce
scores = [45, 78, 23, 89, 56]
max_score = reduce(lambda x, y: x if x > y else y, scores)
print(max_score)
```

```python
from functools import reduce
lst=['mouni','kona','kanchi']
reduce(lambda a,b:a+b,lst)
```

Lambda :A lambda is a small anonymous function in Python.<br>

When you need a quick function for a single operation<br>

It can have any number of inputs, but only one expression<br>

It’s useful for short operations that you don’t need to reuse<br>

Often used with map(), filter(), and reduce()<br>

Saves you from writing a full def function for simple tasks

```python
words = ["apple", "kiwi", "banana"]
words.sort(key=lambda x: len(x))
print(words)  
```

```python
numbers = [1, 2, 3, 4, 5]
new_numbers = list(map(lambda x: x + 10, numbers))
print(new_numbers)
```

filter : It offers a convenient way to filter out all the elements of an iterable , for which the function returns true<br>
filter() is used to select only the items from a list that meet a certain condition.<br>
It doesn’t change the items, it just keeps or removes them.


```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
words = ["apple", "banana", "kiwi", "strawberry", "mango"]
long_words = list(filter(lambda word: len(word) > 5, words))
print(long_words)
```
