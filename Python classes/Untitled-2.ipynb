```python
import requests

# Example: Public API for random jokes
url = "https://official-joke-api.appspot.com/random_joke"

response = requests.get(url)   # send a GET request

# Check status
if response.status_code == 200:
    free = response.json()   # convert JSON response to Python dict
    print("Joke:", free["setup"])
    print("Punchline:", free["punchline"])
else:
    print("Error:", response.status_code)
```

```python
import requests

url = "https://api.agify.io"   # predicts age based on name
params = {"name": "mounika"}

response = requests.get(url, params=params)

print(response.json())  # Example: {'name': 'mounika', 'age': 25, 'count': 1234}
```

```python
import requests

# Get repositories of a GitHub user
url = "https://api.github.com/users/mounika-k/repos"

response = requests.get(url)

if response.status_code == 200:
    repos = response.json()
    for repo in repos[:5]:  # show first 5 repos
        print("Repo Name:", repo["name"])
        print("Stars:", repo["stargazers_count"])
        print("Language:", repo["language"])
        print("---")
else:
    print("Error:", response.status_code)
```
