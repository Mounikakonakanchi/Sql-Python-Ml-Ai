```python
import pandas as pd
credit = pd.read_csv(r"C:\Users\saiko\Downloads\credit_fraud_raw.csv")
print(credit)
```

```python
print(credit.isnull().sum())
```

```python
print(credit.duplicated().sum())
```

```python
credit['Amount'] = pd.to_numeric(credit['Amount'] , errors = 'coerce')
credit = credit.dropna(subset=['Amount'])
```

```python
credit['Timestamp_clean'] = pd.to_datetime(credit['Timestamp'], errors ='coerce')
credit = credit.dropna(subset=['Timestamp_clean'])
```

```python
credit['MerchantCategory'] = credit['MerchantCategory'].replace({'Grcery':'Grocery'})
```

```python
credit = credit.drop_duplicates(subset=['Amount','Timestamp_clean','TransactionID','CardHolderID' ])
```

```python
# Hour of day
credit['Hour'] = credit['Timestamp_clean'].dt.hour

# Day of week
credit['DayOfWeek'] = credit['Timestamp_clean'].dt.day_name()

# Flag outliers
credit['IsOutlier'] = ((credit['Amount'] < 0) | (credit['Amount'] > 5000)).astype(int)
```

```python
credit['cardtoken'] = 'PAN_TOKEN'+credit['CardHolderID']
```

```python
credit.to_csv(r"C:\Users\saiko\Downloads\credit_fraud_cleaned.csv", index=False)
```
