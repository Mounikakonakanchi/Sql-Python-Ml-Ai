For loop : It is used to repeat a block of code multiple times <br>
           It is used when the number of iterations is known

```python
for letter in "hello":
    print(letter)
```

Loop through List

```python
colors = ["red", "blue", "green"]

for c in colors:
    print(c)
```

```python
numbers = [1, 2, 3]    

for i in numbers:
    print(i + 1)
```

For loop with range : generates numbers you can loop over


```python
for i in range(1, 6):        # here it only displays untill 6 but not 6
    print(i)
```

Range with step size

```python
for i in range(3, 30, 2):    # here it gives the output with step size of 2 , it skips every alternative number
 print(i)
```

While loop : It is a control flow statement which allows code to be executed repeatedly depending on whether the condition is true or not  <br>
It is used when the iterations are unkown <br>
It depends on condition


```python
a = True
while a:
    print("Hello!")
    a = False   # Stop the loop after one time
       #i=i+1
```

Count in while loop

```python
i =1
while i < 6:                 
    print(i)
    i+=1    
```

Sum in while loop

```python
i = 1
total = 0
while i <= 5:
    total = total + i                   # Add i to total
    i = i + 1                           # Increase i
print("Sum is:", total)
```

```python
name = ["Mouni" , "maggie", "bhagi", "sai"]

for i in range(len(name)-1, 0, -2):
    print("I like", name[i])
```

Break in While loop : Break statment is used to exit the loop when an external condition is met

```python
i =1
while i<6:
    print(i)
    if i ==3 :       # here the loop will be exited
     break
    i+=1             # This iteration will not be executed
```

Continue in while loop : Continue is used to end the current iteration and executes the next iteration

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue             # It skips 3 and moves to next  number 
    print(i)

```
