Tuple :It is a data type which stores multiple values and it is immutable , ordered and allows duplicates 


```python
a = ('a' , 2 , 'c' , 'a')
print(a)
```

Length of the tuple

```python
a = ('a' , 2 , 'c' , 'a')
print(len(a))
```

Type of the tuple

```python
a = ('a' , 2 , 'c' , 'a')
print(type(a))
```

Accessing the tuple using index

```python
a = ('a' , 2 , 'c' , 'a')
print(a[2])
```

```python
a = ('a' , 2 , 'c' , 'a')
print(a[-2])
```

Count of Tuple

```python
a = ('a' , 2 , 'c' , 'a')
print(a.count('a'))
```

Modifying the tuple : To update or modify tuple we cannot do it directly as it is immutable we need to change the tuple to list and we can update or modify within the list

```python
a = ('a' , 2 , 'c' , 'a')
x = list(a)
print(x)
```

```python
x = ['a', 2, 'c', 'a']
x[2] = '3'
print(x)
```

```python
print(tuple(x))
```

Adding values to the tuple

```python
a = ('a' , 2 , 'c' , 'a')
x = list(a)
print(x)
x.append(4)
print(x)
print(tuple(x))
```

Joining the Tuples

```python
a = ('a' , 2 , 'c' , 'a')
b = (1, 2, 4, 5, 6)
x =a+b
print(x)
```

sets : It is a data type which store multiple values and it is immutable , unordered and does not allow duplicates 

```python
b = {1, 2, 4, 5, 6}
print(b)
```

Accesing item from set : we cannot directly access the values from the set as it is unordered , but we can use for loop

```python
b = {1, 2, 4, 5, 6}
for s in b:
    print(s)
    print(2 in b)
```

Adding items to the set

```python
b = {1, 2, 4, 5, 6}
b.add(8)
print(b)
```

Update set

```python
b = {1, 2, 4, 5, 6}
a= {'a', 'b', 'c','d'}
a.update(b)
print(a)
```

Remove items from the set

```python
b = {1, 2, 4, 5, 6}
b.remove(2)
print(b)
```

Discard set

```python
b = {1, 2, 4, 5, 6}
b.discard(2)
print(b)
```

Clearing the items from the set

```python
b = {1, 2, 4, 5, 6}
b.clear()
print(b)
```

Pop : It removes and returns a random element from a set

```python
b = {1, 2, 3, 4}

a= b.pop()
print(a)
print(b)
```

Cast : convert set to tuple

```python
b = {1, 2, 4, 5, 6}
t = tuple(b)
print(t)
```

Joining set

```python
b = {1, 2, 4, 5, 6}
a = {'a', 'b' ,'c'}
x = a|b
print(x)
x= a.union(b)
print(x)
```

Difference from the set

```python
b = {'a', 2, 4, 5, 6}
a = {'a', 2, 4 ,'c'}
x= b.difference(a)             # it gives all the non matching elements from b
print(x)
x = b-a
print(x)
```

Dictionaries : These are used to store key value pairs , it doesnt allow duplicates , ordered and mutable

```python
a = {'name' : 'mounika' , 'age' : 25 , 'phone' : 345678}
print(a)
```

Accessing Dictionaries

```python
a = {'name' : 'mounika' , 'age' : 25 , 'phone' : 345678}
x=a.keys()
print(x)
```

Conversion of dictionaries

```python
a = {'name' : 'mounika' , 'age' : 25 , 'phone' : 345678}
x=list(a)
print(x)
```

Adding value to the dictionary

```python
a = {'name' : 'mounika' , 'age' : 25 , 'phone' : 345678}
a['id'] = 234
print(a)
```

Subtracting from a dictionary

```python
a = {'name' : 'mounika' , 'age' : 25 , 'phone' : 345678, 'id': 234}
del a['id']
print(a)
```

Nested Dictionaries

```python
students = {
    "Alice": {"age": 20, "grade": "A", "major": "CS"},
    "Bob": {"age": 22, "grade": "B", "major": "Math"},
    "Charlie": {"age": 21, "grade": "A", "major": "Physics"}
}

print(students)
```
